```python
%matplotlib notebook 
#Use 51 instead of 50 to show exact 
#image as explained in lecture
N=51
x= .3+np.zeros(N)
for n in range (N-1):
    x[n+1]= 4*x[n]*(1-(x[n]))
    
#Upodating the inital values of x0:
x1 = .29999+np.zeros(N)

for i in range (N-1):
    x1[i+1]=4*x1[i]*(1-(x1[i]))
    
    
plt.plot(x, color='k',label='X0 = 0.3') 
plt.plot(x1, color='r',label='X0 = 0.29999') 
plt.legend(bbox_to_anchor =(0.69, 1))# option 1)
plt.xlabel(r'iteration $n$', fontsize=18, labelpad =10)
plt.ylabel(r'$x_n$', fontsize=18, labelpad =10)

#This line to change the y axis range


#This is added for remove margins 
plt.margins(0.0)
plt.yticks(np.arange(0, 1.5, 0.5))

plt.show()
```
